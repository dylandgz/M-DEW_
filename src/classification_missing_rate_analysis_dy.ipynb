```python
import pandas as pd
from scipy import stats

# Initialize a dictionary to hold p-values
p_values = {}

# Perform pairwise tests
for i in range(1, 14):
    for j in range(i+1, 14):

        pipeline_i = df.iloc[:, i]
        pipeline_j = df.iloc[:, j]
        # pipeline_j = df[f'pipeline{j}']
        
        # Here you can use the appropriate test, for example `ttest_rel` for a paired t-test
        # or `wilcoxon` for a non-parametric test if your data is not normally distributed.
        t_stat, p_val = stats.ttest_rel(pipeline_i, pipeline_j)
        
        # Store the p-value in the dictionary, keyed by the pipeline pair
        p_values[f'pipeline{i}_vs_pipeline{j}'] = p_val

# Convert the p-values dictionary to a DataFrame for better visualization
p_values_df = pd.DataFrame(list(p_values.items()), columns=['Pipeline Pair', 'P-Value'])

# Display the p-values DataFrame
p_values_df






```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Your existing code here to create the `p_values_df` dataframe...

# Reshaping the DataFrame for heatmap plotting
# We create an empty square DataFrame where both indices and columns represent the pipelines
num_pipelines = 13
p_matrix = pd.DataFrame(index=range(1, num_pipelines+1), columns=range(1, num_pipelines+1))

# Fill the matrix with the p-values
for index, row in p_values_df.iterrows():
    pipeline_pair = row['Pipeline Pair'].split('_vs_')
    i, j = int(pipeline_pair[0].replace('pipeline', '')), int(pipeline_pair[1].replace('pipeline', ''))
    p_matrix.at[i, j] = row['P-Value']
    p_matrix.at[j, i] = row['P-Value']  # symmetric matrix

# Convert the DataFrame to numeric type for proper plotting
p_matrix = p_matrix.astype(float)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(p_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={'label': 'P-Value'})
plt.title('P-Value Heatmap for Pipeline Comparisons')
plt.xlabel('Pipeline')
plt.ylabel('Pipeline')

# Adjust the layout to make sure everything fits
plt.tight_layout()

# Show the plot
plt.show()
```
